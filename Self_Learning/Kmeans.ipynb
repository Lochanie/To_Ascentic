```python
#Generate some artificial data using scikit_learn library (make_blobs is a method)
from sklearn.datasets import make_blobs
```

```python
#create an artificial data
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
#Import KMeans clustering methods to fit the KMeans algorithm
from sklearn.cluster import KMeans
```

```python
#Randomly assigning each observations to cluster find centroids of each cluster
#Kmeans cluster works on two steps-
#Reassign data points for the clusters whose centroids are closer
#Calculate new centroids for each cluster
#Repeat it over and over until centroid varariation closes to each other.
```

```python
#Create an instance from KMEans method
kmeans=KMeans(n_clusters=4)
```

```python
#Fit a model to classify the data
kmeans.fit(data[0])
```

```python
#Get the centers of the clusters in predicted data
kmeans.cluster_centers_
```

```python
#Get the predicted clusters
kmeans.labels_
```

```python
#Compare the results by using a graph
#fig,(ax1,ax2) - in the same figure share the two axis
#plt.subplots(1,2,sharey=True,figsize=(10,6))- 1 horizontal axis, 2 -vertical axis,sharry true-share the same axis,
#fig size-data range from 6 to 10
fig,(ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(10,6))

#plot the original data and predicted data in the same plot
      
ax1.set_title('K means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
      
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')

plt.show()
```
